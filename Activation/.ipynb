```python
import math
```

```python

#* Sigmoid
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
sigmoid(100)
```

```python

#* Tanh
def tanh(x):
    return (math.exp(x) - math.exp(-x))  / (math.exp(x) + math.exp(-x)) 
tanh(-200)
```

```python

#* relu
def relu(x):
    return max(0, x)
relu(-100)
```

```python

#* Leaky relu
def leaky_relu(x):
    return max(0.1*x,x)
leaky_relu(-100)
```
